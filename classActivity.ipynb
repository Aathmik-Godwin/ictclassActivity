```python
file_name = "expenses.txt"


def add_expense():
    try:
        category = input("Enter category: food,shopping,travel,rent ")
        amount = float(input("Enter amount: "))

        file = open(file_name, "a")
        file.write(category + "," + str(amount) + "\n")
        file.close()

        print("Expense added!\n")

    except ValueError:
        print("Invalid amount!\n")

def view_expenses():
    try:
        file = open(file_name, "r")
        print("\nExpenses List:")

        for line in file:
            data = line.strip().split(",")
            print("Category:", data[0], "| Amount:", data[1])

        file.close()
        print()

    except FileNotFoundError:
        print("No file found!\n")


def calculate_total():
    total = 0
    try:
        file = open(file_name, "r")

        for line in file:
            data = line.strip().split(",")
            total += float(data[1])

        file.close()
        print("Total Expense:", total, "\n")

    except FileNotFoundError:
        print("No file found!\n")


while True:
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Total Expense")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_expense()
    elif choice == "2":
        view_expenses()
    elif choice == "3":
        calculate_total()
    elif choice == "4":
        print("Goodbye!")
        break
    else:
        print("Invalid choice!\n")
```
